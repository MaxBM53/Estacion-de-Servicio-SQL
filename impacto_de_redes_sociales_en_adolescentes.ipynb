```python
import pandas as pd
df = pd.read_csv("Social_media_impact_on_life.csv")
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Most_Used_Platform"].value_counts()
```

```python
import matplotlib.pyplot as plt

df['Most_Used_Platform'].value_counts().plot(kind='bar')
plt.title('Plataformas más usadas')
plt.xlabel('Plataforma')
plt.ylabel('Cantidad de estudiantes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df.groupby("Most_Used_Platform")["Sleep_Hours_Per_Night"].mean()
```

```python
df.groupby("Most_Used_Platform")["Sleep_Hours_Per_Night"].mean().sort_values().plot(kind="barh")
plt.title("Promedio de Horas de Sueño por Plataforma")
plt.xlabel("Horas de Sueño")
plt.tight_layout()
plt.show()
```

```python
df.groupby("Affects_Academic_Performance")["Avg_Daily_Usage_Hours"].mean()
```

```python
df.groupby("Affects_Academic_Performance")["Avg_Daily_Usage_Hours"].mean().sort_values().plot(kind="bar")
plt.title("Uso diario promedio según impacto académico")
plt.xlabel("Afecta el rendimiento")
plt.ylabel("Horas diarias")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df.groupby("Overall_Impact")["Mental_Health_Score"].mean().sort_values()
```

```python
df.groupby("Overall_Impact")["Mental_Health_Score"].mean().sort_values().plot(kind="bar")
plt.title("Impacto en la salud mental")
plt.xlabel("Impacto General")
plt.ylabel("Puntaje de salud mental")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
